```
data['invoiceDate']=pd.to_datetime(data['Date'])
data['year']=data['invoiceDate'].dt.year
data['day']=data['invoiceDate'].dt.day
data
```

basket

```
# basket = data.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().fillna(0)
# basket

basket= data.groupby(['year','Item']).sum().unstack().fillna(0)
```

```
basket
```

```
basket=basket.applymap(lambda x:1 if x>0 else 0)
```

```
basket
```

```
popular_sets = apriori(basket, min_support=0.08, use_colnames=True)
```

```
pop_sets
```

```
rules = association_rules(popular_sets,metric='lift',min_threshold=1)  
```
